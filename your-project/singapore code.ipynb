```python
import numpy as np
import pandas as pd
```

```python
occupation_f = pd.read_csv('../datasets/Employed Female Residents Aged 15 Years And Over By Occupation And Age Group.csv')
occupation_m = pd.read_csv('../datasets/Employed Male Residents Aged 15 Years And Over By Occupation And Age Group.csv')
```

```python
print(occupation_f.shape)
print(occupation_m.shape)
```

```python
occupation_f = occupation_f[[' Variables ', ' 2018 ']]
occupation_m = occupation_m[[' Variables ', ' 2018 ']]
```

```python
print(occupation_f.shape)
print(occupation_m.shape)
```

```python
occupation_f = occupation_f.loc[1:11, : ]
occupation_m = occupation_m.loc[1:11, : ]
```

```python
print(occupation_f.shape)
print(occupation_m.shape)
```

```python
occupation_f.columns=['Occupation','Female']
occupation_m.columns=['Occupation','Male']
```

```python
occupation_f
```

```python
occupation_m
```

```python
final = pd.merge(occupation_f, occupation_m, on='Occupation')
final
```

```python
total_f = round(occupation_f['Female'].sum())
print('Total female', total_f)

total_m = round(occupation_m['Male'].sum())
print('Total male', total_m)
```

```python
final['% Female'] = round((final['Female'] / total_f *100),1)
final['% Male'] = round((final['Male'] / total_m *100),1)
final
```

```python
final = final[['Occupation', '% Female', '% Male']]
final
```

```python
final.loc[11] = final.sum()
final
```

```python
final.loc[11, 'Occupation'] = 'Total'
final
```
